```python
# Ensure dependencies are installed
!pip install matplotlib numpy

# Import libraries and set Matplotlib backend
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sys

# Debug: Confirm Matplotlib backend and versions
print(f"Python version: {sys.version}")
print(f"Matplotlib version: {matplotlib.__version__}")
print(f"Numpy version: {np.__version__}")
print(f"Matplotlib backend: {matplotlib.get_backend()}")  # Should be 'module://ipykernel.pylab.backend_inline'

# Graph 1: Bar Chart - Latency Comparison for Whisper-medium
def plot_latency_comparison():
    print("Generating Latency Comparison plot...")
    configurations = ['FP32 (GPU)', 'FP16 (GPU)', '4-bit (GPU)', 'FP32 (CPU)', 'INT8 (CPU)']
    latencies = [1.395, 1.0, 2.7448, 45.3059, 29.2023]  # in seconds

    plt.figure(figsize=(10, 6))
    bars = plt.bar(configurations, latencies, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
    plt.xlabel('Configuration')
    plt.ylabel('Latency (seconds)')
    plt.title('Latency Comparison for Whisper-medium Across Configurations')
    plt.yscale('log')  # Log scale for better visualization
    plt.grid(True, which="both", ls="--", alpha=0.5)

    # Add value labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', va='bottom', ha='center', fontsize=10)

    plt.tight_layout()
    plt.show()
    print("Latency Comparison plot displayed.")

# Graph 2: Pie Chart - Memory Usage for MiniCPM-V 2.6
def plot_memory_usage():
    print("Generating Memory Usage plot...")
    labels = ['4-bit (3.246GB)', 'FP16 (8.972GB)']
    sizes = [3.246, 8.972]  # in GB
    colors = ['#ff9999', '#66b3ff']
    explode = (0.1, 0)  # Explode 4-bit slice

    plt.figure(figsize=(8, 6))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
    plt.title('Memory Usage Comparison for MiniCPM-V 2.6 (4-bit vs. FP16)')
    plt.axis('equal')  # Ensure pie is circular

    plt.tight_layout()
    plt.show()
    print("Memory Usage plot displayed.")

# Graph 3: Line Graph - Throughput vs. Batch Size for DistilGPT-2 and Llama-3-8B
def plot_throughput():
    print("Generating Throughput plot...")
    batch_sizes = [1, 2, 4, 8, 16]
    distilgpt2_throughput = [36.07, 44.54, 92.36, 174.88, 208.97]  # tokens/s
    llama3_throughput = [21.67, 44.53, 92.72, 144.33, 159.81]  # tokens/s

    plt.figure(figsize=(10, 6))
    plt.plot(batch_sizes, distilgpt2_throughput, marker='o', linestyle='-', color='#1f77b4', label='DistilGPT-2')
    plt.plot(batch_sizes, llama3_throughput, marker='s', linestyle='--', color='#ff7f0e', label='Llama-3-8B')
    plt.xlabel('Batch Size')
    plt.ylabel('Throughput (tokens/s)')
    plt.title('Throughput vs. Batch Size for DistilGPT-2 and Llama-3-8B')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()

    # Add data point labels
    for i, txt in enumerate(distilgpt2_throughput):
        plt.annotate(f'{txt:.1f}', (batch_sizes[i], distilgpt2_throughput[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='#1f77b4')
    for i, txt in enumerate(llama3_throughput):
        plt.annotate(f'{txt:.1f}', (batch_sizes[i], llama3_throughput[i]), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8, color='#ff7f0e')

    plt.tight_layout()
    plt.show()
    print("Throughput plot displayed.")

# Graph 4: Bar Chart - TensorRT Speedup for ResNet-50
def plot_tensorrt_speedup():
    print("Generating TensorRT Speedup plot...")
    frameworks = ['PyTorch', 'TensorRT']
    inference_times = [7.25, 1.72]  # in milliseconds

    plt.figure(figsize=(8, 6))
    bars = plt.bar(frameworks, inference_times, color=['#1f77b4', '#ff7f0e'])
    plt.xlabel('Framework')
    plt.ylabel('Inference Time (ms)')
    plt.title('TensorRT Speedup for ResNet-50 Inference')
    plt.grid(True, axis='y', linestyle='--', alpha=0.5)

    # Add value labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', va='bottom', ha='center', fontsize=10)

    plt.tight_layout()
    plt.show()
    print("TensorRT Speedup plot displayed.")

# Execute all plotting functions with error handling
try:
    plot_latency_comparison()
    plot_memory_usage()
    plot_throughput()
    plot_tensorrt_speedup()
except Exception as e:
    print(f"Error generating plots: {e}")
print("All plotting functions executed.")
```

```python
def plot_continuous_batching():
    print("Generating Continuous Batching plot...")
    batch_sizes = [1, 2, 4, 8, 16]
    distilgpt2_throughput = [36.07, 44.54, 92.36, 174.88, 208.97]  # tokens/s
    distilgpt2_latency = [0.36, 0.44, 0.49, 0.53, 0.70]  # seconds
    llama3_throughput = [21.67, 44.53, 92.72, 144.33, 159.81]  # tokens/s
    llama3_latency = [6.92, 6.69, 7.67, 10.32, 18.68]  # seconds

    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Plot throughput on left y-axis
    ax1.plot(batch_sizes, distilgpt2_throughput, marker='o', linestyle='-', color='#1f77b4', label='DistilGPT-2 Throughput')
    ax1.plot(batch_sizes, llama3_throughput, marker='s', linestyle='--', color='#ff7f0e', label='Llama-3-8B Throughput')
    ax1.set_xlabel('Batch Size')
    ax1.set_ylabel('Throughput (tokens/s)', color='black')
    ax1.grid(True, linestyle='--', alpha=0.5)
    ax1.legend(loc='upper left')

    # Plot latency on right y-axis
    ax2 = ax1.twinx()
    ax2.plot(batch_sizes, distilgpt2_latency, marker='^', linestyle='-', color='#2ca02c', label='DistilGPT-2 Latency')
    ax2.plot(batch_sizes, llama3_latency, marker='d', linestyle='--', color='#d62728', label='Llama-3-8B Latency')
    ax2.set_ylabel('Latency (seconds)', color='black')
    ax2.legend(loc='upper right')

    plt.title('Continuous Batching: Throughput and Latency vs. Batch Size')

    # Add data point labels for throughput
    for i, txt in enumerate(distilgpt2_throughput):
        ax1.annotate(f'{txt:.1f}', (batch_sizes[i], distilgpt2_throughput[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='#1f77b4')
    for i, txt in enumerate(llama3_throughput):
        ax1.annotate(f'{txt:.1f}', (batch_sizes[i], llama3_throughput[i]), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8, color='#ff7f0e')

    # Add data point labels for latency
    for i, txt in enumerate(distilgpt2_latency):
        ax2.annotate(f'{txt:.2f}', (batch_sizes[i], distilgpt2_latency[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='#2ca02c')
    for i, txt in enumerate(llama3_latency):
        ax2.annotate(f'{txt:.1f}', (batch_sizes[i], llama3_latency[i]), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8, color='#d62728')

    plt.tight_layout()
    plt.show()
    print("Continuous Batching plot displayed.")
```
